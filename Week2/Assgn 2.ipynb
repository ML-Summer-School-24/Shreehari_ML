```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('winequality-red.csv')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)


df.head()
```

```python
df.isnull().sum()
```

```python
df[df['total sulfur dioxide']>160] 
```

```python

plt.pie(df['quality'].value_counts(), autopct='%0.2f %%', radius = 2, 
        labels = df['quality'].unique(), wedgeprops= {"linewidth":1, "edgecolor":"white"}, 
         pctdistance=0.9)

plt.title('Wine Quality in pie chart', y=1.4) 

#plt.legend(loc="best")

plt.show()
```

```python
plt.title('Wine Quality')

Hist = sns.histplot(data=df, x="quality")
Hist.set_xlabel('Quality')

plt.show()
```

```python
plot =plt.figure(figsize=(5,5))
sns.barplot(x='quality', y='citric acid', data=df)
```

```python
correlation = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True)
plt.title("Correlation Matrix")
plt.show()
```
